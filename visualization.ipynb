```python
data = [1,2,3,4,6,8,70,12]
fig, ax = plt.subplots(figsize=(5,3))
ax.set_xlabel('no of elements', fontdict={'fontsize':10, 'color':'blue'})
ax.set_ylabel('Y - Value')
ax.grid(True)
ax.set_title("Basic of visuals", loc='right')
ax.plot(data)
```

```python
data = [1,2,3,4,6,8,70,12]
fig, ax = plt.subplots(figsize=(5,3))
ax.set_xlabel('no of elements', fontdict={'fontsize':10, 'color':'blue'})
ax.set_ylabel('Y - Value')
ax.grid(True)
ax.set_title("Basic of visuals", loc='right')
ax.plot(data, marker='D', markersize=10, markeredgecolor='black', markerfacecolor='green')
```

```python
data = [1,2,3,4,6,8,70,12]
data2 = [2,5,3,55,33,5,4,65]
fig, ax = plt.subplots(figsize=(5,3))
ax.set_xlabel('no of elements', fontdict={'fontsize':10, 'color':'blue'})
ax.set_ylabel('Y - Value')
ax.grid(True)
ax.set_title("Basic of visuals", loc='right')
ax.plot(data, marker='D', markersize=10, markeredgecolor='black', markerfacecolor='green', linestyle=':')
ax.plot(data2, marker='X', markersize=10, markeredgecolor='black', markerfacecolor='red', linestyle='--', color='blue', linewidth=3, alpha=0.5)
```

```python
data = [1,2,3,4,6,8,70,12]
data2 = [2,5,3,55,33,5,4,65]
fig, (ax1, ax2) = plt.subplots(figsize=(10,3), ncols=2)
ax1.plot(data, marker='D', markersize=10, markeredgecolor='black', markerfacecolor='green', linestyle=':', linewidth=3)
ax2.plot(data2, marker='X', markersize=10, markeredgecolor='black', markerfacecolor='red', linestyle='--', color='blue', linewidth=3)
```
